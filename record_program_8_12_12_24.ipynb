```python
#Rectangle class with Area Comparison
class Rectangle:
    def get(self, l, b):
        self.l = l
        self.b = b
        
    def area(self):
        return self.l * self.b
    
    def perimeter(self):
        return 2 * (self.l + self.b)

r1 = Rectangle()
r2 = Rectangle()

l1 = int(input("Enter length of Rectangle 1: "))
b1 = int(input("Enter breadth of Rectangle 1: "))
r1.get(l1, b1)

l2 = int(input("Enter length of Rectangle 2: "))
b2 = int(input("Enter breadth of Rectangle 2: "))
r2.get(l2, b2)

ar1 = r1.area()
ar2 = r2.area()

if ar1 > ar2:
    print("Rectangle 1 has greater area")
elif ar1 < ar2:
    print("Rectangle 2 has greater area")
else:
    print("Both rectangles have the same area")
```

```python
#bank account class with deposit and withdrawal
class BankAccount:
    def __init__(self, a_name, a_no, a_type, a_balance):
        self.acc_name = a_name
        self.acc_no = a_no
        self.acc_type = a_type
        self.acc_balance = int(a_balance) 
    
    def deposit(self, a_deposit):
        print("Initial balance:", self.acc_balance)
        self.acc_balance += a_deposit
        print(f"{a_deposit} is deposited")
        print("Current balance:", self.acc_balance)
        
    def withdraw(self):
        try:
            self.amount = int(input("Withdraw amount: "))
            print("Initial balance:", self.acc_balance)
            if self.amount > self.acc_balance:
                print("You don't have enough balance to withdraw!")
            else:
                self.acc_balance -= self.amount
                print(f"{self.amount} is withdrawn")
                print("Current balance:", self.acc_balance)
        except ValueError:
            print("Please enter a valid amount to withdraw.")
            
    def acc_info(self):
        print("Account holder name:", self.acc_name)
        print("Account number:", self.acc_no)
        print("Account type:", self.acc_type)
        print("Account balance:", self.acc_balance)
        
name = input("Enter name: ")
no = input("Enter account number: ")
acc_type = input("Enter account type: ")
bal = input("Enter initial balance: ")

holder = BankAccount(name, no, acc_type, bal)

while True:
    ch = int(input("\n1. Deposit\n2. Withdraw\n3. Account Info\n4. Exit\nChoose your option: "))
    if ch == 1:
        try:
            amount = int(input("Deposit amount: "))
            holder.deposit(amount)
        except ValueError:
            print("Please enter a valid deposit amount.")
    elif ch == 2:
        holder.withdraw()
    elif ch == 3:
        holder.acc_info()
    elif ch == 4:
        print("Exiting...")
        break
    else:
        print("Invalid option. Please try again.")
```

```python
#rectangle class with operator overloading
class Rectangle:
    def __init__(self, l, w):
        self.length = l
        self.width = w
        self.area = l * w 
    
    def __gt__(self, other):
        if self.area > other.area:
            print("Rectangle 1 is greater")
        else:
            print("Rectangle 2 is greater") 
            
l1 = int(input("Enter length of Rectangle 1: "))
w1 = int(input("Enter width of Rectangle 1: ")) 
l2 = int(input("Enter length of Rectangle 2: ")) 
w2 = int(input("Enter width of Rectangle 2: ")) 

r1 = Rectangle(l1, w1)
r2 = Rectangle(l2, w2) 

r1 > r2
```

```python
#time class with + operator overloading
class Time: 
    def __init__(self, hr, mint, sec):
        self.__h = hr
        self.__m = mint
        self.__s = sec
        
    def __add__(self, other):
        hh = self.__h + other.__h
        mm = self.__m + other.__m
        ss = self.__s + other.__s
        
        if ss >= 60:
            mm += ss // 60
            ss = ss % 60
            
        if mm >= 60:
            hh += mm // 60
            mm = mm % 60
            
        print(f"Added Time: {hh}h {mm}m {ss}s") 
        
h1 = int(input("Enter hours for time1: "))
m1 = int(input("Enter minutes for time1: "))
s1 = int(input("Enter seconds for time1: "))

h2 = int(input("Enter hours for time2: "))
m2 = int(input("Enter minutes for time2: "))
s2 = int(input("Enter seconds for time2: ")) 

t1 = Time(h1, m1, s1)
t2 = Time(h2, m2, s2) 

t1 + t2
```

```python
#publisher,book and python book class
class Publisher:
    def __init__(self, pname):
        self.pubname = pname
        
    def display1(self):
        print("Book Details")
        print("Publisher Name:", self.pubname)
        
class Book(Publisher):
    def __init__(self, pname, bname, author):
        super().__init__(pname)
        self.title = bname
        self.auth = author
        
    def display2(self):
        print("Title:", self.title)
        print("Author:", self.auth)
        
class Python(Book):
    def __init__(self, pname, title, au, price, np):
        super().__init__(pname, title, au)
        self.rate = price
        self.nop = np
        
    def display3(self):
        print("Rate:", self.rate)
        print("No. of Pages:", self.nop)
        
p = input("Enter publisher name: ")
b = input("Enter book name: ")
a = input("Enter author name: ")
r = int(input("Enter rate of book: "))
n = int(input("Enter number of pages: ")) 

python_book = Python(p, b, a, r, n)
python_book.display1()
python_book.display2()
python_book.display3()
```
